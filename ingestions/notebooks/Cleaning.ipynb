```python
# Load the SUDORS data
try:
    df_sudors = pd.read_csv("SUDORS-Fatal-Overdose-Data.csv")
except FileNotFoundError:
    raise FileNotFoundError("Error: 'SUDORS-Fatal-Overdose-Data.csv' not found. Please ensure the file is uploaded.")

# Define the columns for aggregation.
columns_to_aggregate_user_input = [
    'opioids_rate',
    'rx_drugs_percent',
    'illegal_drugs_percent',
    'counterfeit_pills_percent',
    'homeless_percent',
    'curr_paintrt_percent',
    'ever_MHSUDtrt_percent',
    'curr_MHSUDtrt_percent',
    'hx_selfharm_percent',
    'recentopioid_percent',
    'priorod_mon_percent',
    'intervopp_percent',
    'curr_SUDtrt_percent',
    'druguse_witnessed_percent',
    'mhdiag_percent',
    'bystander_percent',
    'priorod_percent',
    'recentrelease_percent' # Assuming 'recentrelease' was meant to be 'recentrelease_percent'
]

# Identify the actual 'YEAR' and 'STATE' columns.
# Common column names for state and year in SUDORS data
state_col_options = ['State', 'STATE', 'state', 'JURISDICTION', 'Jurisdiction'] # Added 'Jurisdiction'
year_col_options = ['Year', 'YEAR', 'year']

state_col = None
year_col = None

# Check for state and year column names
for col in state_col_options:
    if col in df_sudors.columns:
        state_col = col
        break

for col in year_col_options:
    if col in df_sudors.columns:
        year_col = col
        break

if not state_col:
    raise ValueError("Error: Could not find a 'State' column. Please check the column names in 'SUDORS-Fatal-Overdose-Data.csv'.")

if not year_col:
    raise ValueError("Error: Could not find a 'Year' column. Please check the column names in 'SUDORS-Fatal-Overdose-Data.csv'.")

# Filter for existing columns to aggregate
existing_columns_to_aggregate = [
    col for col in columns_to_aggregate_user_input if col in df_sudors.columns
]

# Add a warning if some columns were not found
missing_columns = [
    col for col in columns_to_aggregate_user_input if col not in df_sudors.columns
]
if missing_columns:
    print(f"Warning: The following requested columns were not found in the dataset: {', '.join(missing_columns)}. They will be skipped.")

if not existing_columns_to_aggregate:
    raise ValueError("Error: None of the specified columns for aggregation were found in the dataset. Please check the column names.")

# Combine grouping columns and aggregation columns
selected_columns = [year_col, state_col] + existing_columns_to_aggregate

# Filter the DataFrame to include only the selected columns
df_filtered = df_sudors[selected_columns].copy()

# Ensure numeric types for aggregation columns, coercing errors to NaN
for col in existing_columns_to_aggregate:
    df_filtered[col] = pd.to_numeric(df_filtered[col], errors='coerce')

# Drop rows where all aggregation columns are NaN after conversion to avoid issues with averaging
df_filtered.dropna(subset=existing_columns_to_aggregate, inplace=True)


# Step 1: Average data from each year for each state
# This involves grouping by year and state and calculating the mean for the value columns
df_yearly_state_avg = df_filtered.groupby([year_col, state_col])[existing_columns_to_aggregate].mean().reset_index()

# Step 2: Average the data from each year for each state to one data point per state
# This involves grouping the result from Step 1 by state and calculating the mean
# across years for each state.
df_state_avg = df_yearly_state_avg.groupby(state_col)[existing_columns_to_aggregate].mean().reset_index()

# Create a CSV called circumstance
csv_filename = "circumstance.csv"
df_state_avg.to_csv(csv_filename, index=False)

print(f"Successfully created '{csv_filename}' with averaged data per state.")
print("Here's a preview of the created CSV:")
print(df_state_avg.head())
```
