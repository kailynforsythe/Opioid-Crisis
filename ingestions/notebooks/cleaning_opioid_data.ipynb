```python
import pandas as pd
df = pd.read_csv("nsduh.csv")
df.head()
```

```python
columns_to_filter = [
    'SERVICE', 'MILSTAT', 'NOMARR2', 'LVLDIFMEM2', 'LVLDIFCARE2', 'IRSEX',
    'IREDUHIGHST2', 'SEXAGE', 'HEALTH2', 'EDUSCKMON', 'EDUSKPMON', 'WRKDRGPOL',
    'WRKDRGEDU', 'WRKTSTALC', 'WRKTSTRDM', 'WRKOKPREH', 'WRKOKRAND', 'WRKSTATWK2',
    'IMOTHER', 'IFATHER', 'IRHH65_2', 'IRFAMSOC', 'IRFAMSSI', 'IRFSTAMP',
    'IRFAMPMT', 'IRFAMSVC', 'IRPINC3', 'GOVTPROG', 'BOOKED', 'BKSRVIOL',
    'BKSMASLT', 'BKPOSTOB', 'BKDRVINF', 'BKDRUG', 'PNRNMLAS1', 'OPINMMON', 'FUPNRNM18', 'OPINMYR'
]

# Filter the DataFrame to include only the specified columns
df_filtered = df[columns_to_filter]
df_opinmyr = df_filtered[df_filtered['OPINMYR'] == 1]
df_opinmyr.head()

# Save the filtered DataFrame to a CSV file
df_filtered.to_csv('filtered_nsduh.csv', index=False)

print("Filtered data saved to 'filtered_nsduh.csv'")
df_filtered.head()
```
