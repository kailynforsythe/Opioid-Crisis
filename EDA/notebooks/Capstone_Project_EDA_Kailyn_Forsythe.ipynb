```python
import matplotlib.pyplot as plt
import os
os.makedirs("graphs", exist_ok=True)

# Filter data for 2024
sudors_2024 = sudors[sudors['year'] == 2024]

# Calculate average opioid percent and rx opioid percent
avg_opioids_percent = sudors_2024['opioids_percent'].mean()
avg_rx_opioids_percent = sudors_2024['rx_OUD_percent'].mean()

# Create a bar graph
labels = ['Average Opioid Percent', 'Average Rx Opioid Percent']
values = [avg_opioids_percent, avg_rx_opioids_percent]

plt.figure(figsize=(8, 6))
plt.bar(labels, values, color=['skyblue', 'lightcoral'])
plt.ylabel('Percentage')
plt.title('Average Opioid and Rx Opioid Percentages in 2024')
plt.ylim(0, max(values) * 1.2) # Set y-axis limit for better visualization
plt.savefig("Average Opioid and Rx Opioid Percentages in 2024.png", bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the percentage columns for the specified drugs
drug_percent_columns = [
    'carfentanil_percent',
    'parafluorofentanyl_percent',
    'other_illegal_opioids_percent',
    'nitazenes_percent',
    'isotonitazene_percent',
    'metonitazene_percent'
]

# Calculate the average percentages for each drug for the year 2024
# Convert columns to numeric, coercing errors to NaN and then filling NaN with 0
average_drug_percents = sudors_2024[drug_percent_columns].apply(pd.to_numeric, errors='coerce').fillna(0).mean()

# Prepare data for plotting
labels_percents = [
    'Carfentanil',
    'Parafluorofentanyl',
    'Other Illegal Opioids',
    'Nitazenes',
    'Isotonitazene',
    'Metonitazene'
]
values_percents = average_drug_percents.values

plt.figure(figsize=(12, 7))
plt.bar(labels_percents, values_percents, color=['skyblue', 'lightcoral', 'lightgreen', 'orange', 'purple', 'yellowgreen'])
plt.ylabel('Average Percentage')
plt.title('Average Percentage of Deaths by Specific Opioid Drugs in 2024')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, max(values_percents) * 1.2) # Set y-axis limit for better visualization
plt.tight_layout()
plt.savefig("Average Percentage of Deaths by Specific Opioid Drugs in 2024.png", bbox_inches="tight")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Calculate the average opioid death rate per state from the SUDORS data
average_opioid_rate_by_state = sudors.groupby('Jurisdiction')['opioids_rate'].mean().reset_index()

# Sort states by opioid death rate in descending order and select the top 10
top_10_opioid_states = average_opioid_rate_by_state.sort_values(by='opioids_rate', ascending=False).head(10)

# Calculate the average Naloxone Dispensing Rate per state from the naloxone data
average_naloxone_rate_by_state = naloxone.groupby('Location')['Naloxone Dispensing Rate'].mean().reset_index()

# Merge the top opioid states data with the average naloxone dispensing rates
merged_data = pd.merge(top_10_opioid_states, average_naloxone_rate_by_state, left_on='Jurisdiction', right_on='Location', how='inner')

# Drop the duplicate 'Location' column as 'Jurisdiction' is sufficient
merged_data = merged_data.drop(columns='Location')

# Prepare data for plotting
states = merged_data['Jurisdiction']
opioid_rates = merged_data['opioids_rate']
naloxone_rates = merged_data['Naloxone Dispensing Rate']

# Scale naloxone dispensing rates for better visualization
scaling_factor = 50
scaled_naloxone_rates = naloxone_rates * scaling_factor

x = np.arange(len(states))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(14, 8))
rects1 = ax.bar(x - width/2, opioid_rates, width, label='Average Opioid Death Rate', color='skyblue')
rects2 = ax.bar(x + width/2, scaled_naloxone_rates, width, label=f'Average Naloxone Dispensing Rate (per 100 persons) x{scaling_factor}', color='lightcoral')

# Add labels, title, and custom x-axis tick labels
ax.set_ylabel('Rate')
ax.set_title('Top 10 States by Opioid Death Rate: Opioid Death Rate vs. Scaled Naloxone Dispensing Rate')
ax.set_xticks(x)
ax.set_xticklabels(states, rotation=45, ha='right')
ax.legend()
ax.set_ylim(0, max(max(opioid_rates), max(scaled_naloxone_rates)) * 1.2) # Set y-axis limit for better visualization

fig.tight_layout()
plt.savefig("Top Opioid Death Rate States - Opioid vs Scaled Naloxone Rates.png", bbox_inches="tight")
plt.show()
```
